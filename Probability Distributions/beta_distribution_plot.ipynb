We learn the beta distribution and plot a few distributions by changing the values of a and b, to get a feel of hyperparameter tuning.

```python
from scipy.special import gamma
def beta(mu, a, b):
    # print(f"[{mu}, {a}, {b}]")
    if((mu == 0 or mu == 1) and (a < 1 or b < 1)):
        return 3
    return (gamma(a + b) / (gamma(a) * gamma(b))) * (mu ** (a - 1)) * ((1 - mu) ** (b - 1))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python

def plot_beta_distribution(a, b):
    fig, ax = plt.subplots()
    xs = []
    ys = []
    for i in range(100):
        xs.append((i / 100.0))
        ys.append(beta((i / 100.0), a, b))
    ax.plot(xs, ys)
    plt.plot()
```

```python
plot_beta_distribution(2, 3)
```
